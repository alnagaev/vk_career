```python
import requests

token = '2bcea6c17817bb930faa394ca43ab164483205cc'# сюда вставляем токен 
headers = {'Authorization': 'token ' + token}

user = 'discospiff' #расскоментить для тестирования

# user = str(input()) # закоментить при тестировании
r = requests.get('https://api.github.com/users/{}'.format(user), headers=headers) 
res = r.json()
res
```

```python
#Основная функция. В качестве аргумента принимает как одного юзера, так и массив
#Результат - эмейл и параметр hireable(None - не ищет работу, True - ищет)
def get_email(user, *users):
    if len(users) > 1:
        dct = {}
        for user in users:
                try:
                    r = requests.get('https://api.github.com/users/{}'.format(user), headers=headers)
                    res = r.json()
                    mail = res['email']
                    hireable = res['hireable']
                    dct[user] = (mail, hireable)
                except Exception as e:
                    print(str(e))
        return dct

    r = requests.get('https://api.github.com/users/{}'.format(user), headers=headers)
    res = r.json()
    mail = res["email"]

#     return ("Эмейл: {}".format(mail), "Ищет ли работу в данный момент: {}".format(hireable)) красивый принт
    return (mail, hireable)
      
print(get_email('discospiff'))
```

```python
#получаем репозитории по ключевому запросу
query = 'java_enterprise'
# query = str(input())

r = requests.get("https://api.github.com/search/repositories?q={}".format(query), headers=headers)
res = r.json()
res
```

```python
#Из json объекта извлекаем нужный нам урл, чтобы получить список девелоперов репозитория
url = res['items'][0]['archive_url'].split('{')[0]
url
```

```python
contributors = requests.get('{}contributors'.format(url), headers=headers).json()
contributors
```

```python
#получили их ники, теперь можно запускать функцию
contr = [contributors[i]['login'] for i in range(len(contributors))]
contr
```

```python
import time
start_time = time.time()
for i in contr:
    print(i, get_email(i))
    
print("--- %s seconds ---" % (time.time() - start_time))
```
